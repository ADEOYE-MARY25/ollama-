# Creating a range and tensors like

```python
# Sometimes you want a range of numbers such as 1 to 10 or 0 to 100
# you can (start, end , step) 

# Use torch.arange(), torch.range() is deprecated
zero_to_ten_deprecated = torch.range(0, 10) # Note: this may return an error in the future

# Create  a range of values 0 to 10
zeros_to_ten = torch.arange(start=0, end=10, step=1)
zeros_to_ten
```

```python
# sometimes you might want one tensor of certain type with the same shape as another tensor

# can also create a tensor of zeroes similar to another tensor
ten_zeros = torch.zeros_like(input=zeros_to_ten) # will have same shape
ten_zeros
```

# Manipulating tensors (tensor operations)

```python
# Basic operations
# Create a tensor of values and add a number to it
tensor = torch.tensor([1,2,3])
tensor + 10
```

```python
# Multiply it by 10
tensor * 10
```

```python
# Tensors don't change unless reassigned
tensor
```

```python
# Subtract and reassign
tensor = tensor - 10
tensor
```

```python
# Add and reassign
tensor = tensor + 10
tensor
```

```python
# Can also use torch functions
torch.multiply(tensor, 10)
```

```python
# Original tensor is still unchanged
tensor
```

```python
# Element-wise multiplication (each element multiplies its equivalent, index 0->0, 1 ->1, 2 ->2)
print(tensor, "*", tensor)
print("Equals:",  tensor * tensor)
```

# Matrix multiplication (is all you need)

```python
import torch
tensor = torch.tensor([1, 2, 3])
tensor.shape
```

```python
# Element-wise matrix multiplication
tensor * tensor
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
# Can also use the "@" symbol for matrix multiplication, though not recommended
tensor @ tensor
```

```python
%%time
# Matrix multiplication by hand
# (avoid doing operations with for loops at all cost, they are computationally expensive)
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
value
```

```python
%%time
torch.matmul(tensor, tensor)
```

# One of the most common errors in deep learning (shape errors)

```python
# Shape need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)
                         
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)
torch.matmul(tensor_A, tensor_B) # (this will error)                         
```

```python
# View tensor_A and tensor_B
print(tensor_A)
print(tensor_B)
```

```python
# View tensor_A and tensor_B.T
print(tensor_A)
print(tensor_B.T)
```
